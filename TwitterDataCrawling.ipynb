```
from twitterscraper.query import query_tweets
from IPython.display import display
import pandas as pd
import datetime

tweets = []
keyword = 'K리그'
cols = ['timestamp','text']

list_of_tweets = query_tweets(keyword, begindate=datetime.date(2018,2,1), enddate=datetime.date(2020,2,2), limit=10)
for tweet in list_of_tweets:
    tweets.append([tweet.timestamp,tweet.text])
    
data_frame = pd.DataFrame(tweets,columns=cols)
display(data_frame)
data_frame.to_csv('/content/gdrive/My Drive/Colab Notebooks/DataCrawling/keyword+3.csv', mode='w', encoding="utf-8-sig")
```
